# 예시 데이터셋으로 sklearn의 swill roll을 3차원으로 가져와보겠습니다.

```python
from sklearn.datasets import make_swiss_roll
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Swiss roll 데이터 생성
X, color = make_swiss_roll(n_samples=1500)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.view_init(10, -80)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=color, cmap=plt.cm.Spectral, edgecolor='k')

plt.show()
```

# PCA를 때리면 어떻게 될까요?

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_r = pca.fit_transform(X)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(X_r[:, 0], X_r[:, 1], c=color, cmap=plt.cm.Spectral, edgecolor='k')
plt.show()
```

# 자 보셨죠? PCA는 그대로 차원을 내리기만 하네요. 뻔하죠...
# 그렇다면 우리의 t-SNE는요?

```python
from sklearn.manifold import TSNE
# tsne = TSNE(n_components=2, perplexity=30)
tsne = TSNE(n_components=2, perplexity=50)
X_r = tsne.fit_transform(X)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(X_r[:, 0], X_r[:, 1], c=color, cmap=plt.cm.Spectral, edgecolor='k')
plt.show()
```

# 이야~ 데이터의 익힘 정도가 다르죠. 정말 Even하게 시각화되네요.

# 숫자 데이터는 어떨까요?

```python
from sklearn.datasets import load_digits
digits = load_digits()

from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

# digits.data.shape
digits.target
```

# PCA를 조져보시죠.

```python
X = digits.data
y = digits.target

X_pca = PCA(n_components=2).fit_transform(X)
pca_x = X_pca[:,0]
pca_y = X_pca[:,1]

fig = plt.figure()
ax = fig.add_subplot()
ax.scatter(pca_x, pca_y, c=y)
```
